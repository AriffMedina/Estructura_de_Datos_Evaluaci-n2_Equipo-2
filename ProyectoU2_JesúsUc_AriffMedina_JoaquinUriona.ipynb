```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, x):
        self.items.append(x)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("La cola está vacía")
        return self.items.pop(0)

    def first(self): # o peek()
        return self.items[0] if not self.is_empty() else None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


class Graph:
    def __init__(self, directed: bool = False):
        self.directed = directed
        self.adj = {}  # dict: vertex -> set(neighbors)

    def add_vertex(self, v):
        if v not in self.adj:
            self.adj[v] = set()

    def add_edge(self, u, v):
        self.add_vertex(u)
        self.add_vertex(v)

        self.adj[u].add(v)
        if not self.directed:
            self.adj[v].add(u)

    def neighbors(self, v):
        return self.adj.get(v, set())

    def __str__(self):
        lines = []
        for v in self.adj:
            nbrs = ", ".join(map(str, sorted(self.adj[v])))
            lines.append(f"{v}: [{nbrs}]")
        return "\n".join(lines)

    #VERIFICAR ARISTAS
    def has_edge(self, u, v):
      if v in self.adj[u] and u in self.adj[v]:
        return True
      else:
        return False

    #ELIMINAR ARISTAS
    def remove_edge(self, u, v):
      if self.has_edge(u, v):
        self.adj[u].remove(v)
        if not self.directed:
          self.adj[v].remove(u)
        return None
      else:
        raise ValueError("no se puede eliminar (u en v) porque no se conectan.")
        # se hace la validacion porque si no tengo las aristas u en v se va a trabar el programa o puede generar algun conflicto


    def remove_vertex(self, v):
      if v not in self.adj:
        raise ValueError("el vertice v no esta en el grafo o el programador lo programo mal ;)")
        # se hace la validacion porque si no tengo V, como se va a eliminar V y puede ser que se pueda generar algun conflicto

      for vertex in list(self.adj.keys()):
        if v in self.adj[vertex]:
          self.adj[vertex].remove(v)

      del self.adj[v]

    def degree(self, v) -> int:
      if v not in self.adj:
        raise ValueError("el vertice v no esta en el grafo.")
        #se hace la validacion porque si no hay V, como se va a calcular y es para que no siga ejecutando lineas de codigo o pueda haber algun fallo
      return len(self.adj[v])

    def out_degree(self, v) -> int:
      if v not in self.adj:
        raise ValueError("el vertice v no esta en el grafo.")
        #se hace la validacion porque si no hay V, como se va a calcular y es para que no siga ejecutando lineas de codigo o pueda haber algun fallo
      return len(self.adj[v])

    def in_degree(self, v) -> int:
      if v not in self.adj:
        raise ValueError("el vertice v no esta en el grafo.")
        #se hace la validacion porque si no hay V, como se va a calcular y es para que no siga ejecutando lineas de codigo o pueda haber algun fallo
      contador = 0
      for vertex_key in self.adj:
        if v in self.adj[vertex_key]:
          contador += 1
      return contador



    def bfs(self, start):
        if start not in self.adj:
            return []

        visited_order = []
        q = Queue()
        discovered = {start}

        q.enqueue(start)

        while not q.is_empty():
            v = q.dequeue()
            visited_order.append(v)

            # Usamos sorted para que el orden sea predecible en las pruebas
            for neighbor in sorted(self.adj[v]):
                if neighbor not in discovered:
                    discovered.add(neighbor)
                    q.enqueue(neighbor)
        return visited_order

    # 4. BFS Parents: Retorna diccionario para reconstruir caminos
    def bfs_parents(self, start):
        if start not in self.adj:
            return {}

        parents = {start: None}
        q = Queue()
        q.enqueue(start)

        while not q.is_empty():
            v = q.dequeue()
            for neighbor in sorted(self.adj[v]):
                if neighbor not in parents:
                    parents[neighbor] = v
                    q.enqueue(neighbor)
        return parents

    # 5. Path Unweighted: Retorna la ruta más corta
    def path_unweighted(self, start, goal):
        if start not in self.adj or goal not in self.adj:
            return None

        parents = self.bfs_parents(start)

        if goal not in parents:
            return None # No hay camino

        # Reconstrucción del camino desde el final al inicio
        path = []
        current = goal
        while current is not None:
            path.append(current)
            current = parents[current]

        return path[::-1]
```

```python
g = Graph(directed=False)
edges = [("A", "B"), ("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"), ("C", "E"), ("D", "E")]
for u, v in edges:
    g.add_edge(u, v)

print(g)
```
